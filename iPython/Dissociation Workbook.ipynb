```python
import os
import numpy as np
```

```python
files1 = os.listdir('To_Plot/RD_vs_Thickness/RD_aft_Ext_Top_Port_07-24-2017/')
files2 = os.listdir('To_Plot/RD_vs_Thickness/Thickness_vs_RD_after_Extension_bottom_Port/')
files3 = os.listdir('To_Plot/RD_vs_Thickness/Thickness_vs_RD_before_Extension/')
```

```python
x = [14.48, 14.23, 18.45, 19.41]
x3 = [20.12, 18.6, 25.23, 26.81]
y1 = []
y2 = []
y3 = []
for f in range(0, len(files1)):
    sdf1 = dscn.SpectrumDataFile('To_Plot/RD_vs_Thickness/RD_aft_Ext_Top_Port_07-24-2017/' + files1[f])
    sdf1.parse_spectral_data()
    sdf1.calculate_dissociation()
    y1.append(sdf1.dissociation)
for f in range(0, len(files2)):
    print files2[f]
    sdf2 = dscn.SpectrumDataFile('To_Plot/RD_vs_Thickness/Thickness_vs_RD_after_Extension_bottom_Port/' + files2[f])
    sdf2.parse_spectral_data()
    sdf2.calculate_dissociation()
    y2.append(sdf2.dissociation)
for f in range(0, len(files3)):
    sdf3 = dscn.SpectrumDataFile('To_Plot/RD_vs_Thickness/Thickness_vs_RD_before_Extension/' + files3[f])
    sdf3.parse_spectral_data()
    sdf3.calculate_dissociation()
    y3.append(sdf3.dissociation)
```

```python
plt.figure()
plt.title('RD vs Thickness')
plt.ylabel('Relative Dissociation')
plt.xlabel('Thickness (A)')
c1 = plt.plot(x, y1, '^', label='Top Port After Inserts')
#c2 = plt.plot(x, y2, '^', label='Bottom Port')
c3 = plt.plot(x3, y3, '^', label='Top Port Before Inserts')
p1 = np.poly1d(np.polyfit(x, y1, 1))
p2 = np.poly1d(np.polyfit(x, y2, 1))
p3 = np.poly1d(np.polyfit(x3, y3, 1))
xp = np.linspace(np.min(x), np.max(x), 100)
xp3 = np.linspace(np.min(x3), np.max(x3), 100)
plt.plot(xp, p1(xp), '--', label='Top Port After Insters, Fit', color=c1[0].get_color())
#plt.plot(p2(xp), xp, '--', label='Bottom Port Fit', color=c2[0].get_color())
plt.plot(xp3, p3(xp3), '--', label='Top Port Before Inserts, Fit', color=c3[0].get_color())
xticks, xticklabels = plt.xticks()
xmin = (3*xticks[0] - xticks[1])/2.
xmax = (3*xticks[-1] - xticks[-2])/2.
plt.xlim(xmin, xmax)
plt.xticks(xticks)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()
```
